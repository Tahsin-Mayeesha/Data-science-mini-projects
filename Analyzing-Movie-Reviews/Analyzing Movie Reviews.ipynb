#  Metacritique vs Fandango scoring



<i>Why would the median for Metacritic_norm_round be lower than the mean, 
but the median for Fandango_Stars is higher than the mean? </i>

The mean is usually larger than the median when there are a few large values in the data, 
and lower when there are a few small values. There are many 5 star movies in Fandango rating and the lowest rated movie is at least 3, so the mean is higher than the median in Fandango ratings. But there are many lower values in Metacritic dataset which makes the mean of metacritic lower than median.

<i>Why would the standard deviation for Fandango_Stars be much lower than the standard deviation for Metacritic_norm_round?</i>

The data for Fandango is very centered and between 3.0 to 5.0, but the metacritic ratings vary from 0.5 to 4.5. Metacritic data is also more scattered.

3.<i>Why would the mean for Fandango_Stars be much higher than the mean for Metacritic_norm_round.?</i>

Fandango ratings are clustered around 4.0 and 4.5 while Metacritique has many smaller scores from 0.5 to 3. This drags the metacritic mean down but perhaps is a better reflection of reality.

```python
from scipy.stats import linregress

slope, intercept, r_value, p_value, stderr_slope = linregress(movies["Metacritic_norm_round"], movies["Fandango_Stars"])
```

```python
pred = slope*movies["Metacritic_norm_round"] + intercept
```

```python
x_values = [3.0,4.0]
y_values = []
for item in x_values:
    prediction = slope*item + intercept
    y_values.append(prediction)
    
plt.plot(x_values,y_values)
```

```python
plt.plot(movies["Metacritic_norm_round"],pred)
plt.plot(movies["Metacritic_norm_round"],movies["Fandango_Stars"],"ro")
```

```python
error = movies["Fandango_Stars"]-pred
```
